```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Malini\Downloads\Mall_Customers.csv")
```

```python
df
```

```python
df['Gender']=df['Gender'].replace(['Male','Female'],[0,1])
print(df.head(5))
```

```python
#label encoding

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df.head()
```

```python
#dendrogram data visualization

import scipy.cluster.hierarchy as clus
plt.figure(1,figsize = (16,8))
dendrogram = clus.dendrogram(clus.linkage(df,method = "ward"))

plt.title("Dendrogram Tree Graph")
plt.xlabel('Customer')
plt.ylabel("Distance")
plt.show()
```

```python
#fitting the hierarchical clustering into the dataset with n=5

from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage="average")
y_means = model.fit_predict(df)
y_means
```

```python
x = df.iloc[:,[3,4]].values
plt.scatter(x[y_means == 0,0],x[y_means == 0,1],s = 50,c = '#92f7ad',label = 'cluster1')
plt.scatter(x[y_means == 1,0],x[y_means == 1,1],s = 50,c = '#f7f292',label = 'cluster2')
plt.scatter(x[y_means == 2,0],x[y_means == 2,1],s = 50,c = '#92eff7',label = 'cluster3')
plt.scatter(x[y_means == 3,0],x[y_means == 3,1],s = 50,c = '#cb92f7',label = 'cluster4')
plt.scatter(x[y_means == 4,0],x[y_means == 4,1],s = 50,c = '#f792d6',label = 'cluster5')
plt.title('income spend analysis')
plt.xlabel('income')
plt.ylabel('spend')
plt.show()
```

# PCA ALGORITHM

```python
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
df= datasets.load_iris()
```

```python
#data seggragation
x=df.data
y=df.target
names=df.target_names
print(x)
print(y)
print(names)
```

```python
#FITTING the pca algorithm to the dataset with n=2

from sklearn.decomposition import PCA
model=PCA(n_components=2) #number of components to keep
y_means =  model.fit(x).transform(x)
print(y_means)
```

```python
plt.figure()
colors=['#84bfe0','#f5f390','#edaaf0']
for color,i,target_name in zip(colors,[0,1,2],names):
    plt.scatter(y_means[y == i,0],y_means[y==i,1],color=color,lw=2,label=target_name)
plt.title('IRIS CLUSTERING')
plt.show()
    
```
